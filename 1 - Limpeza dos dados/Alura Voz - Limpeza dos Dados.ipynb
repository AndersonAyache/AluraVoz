```python
# Importando o pandas para fazer a análise 
import pandas as pd

# Importando o pacote pessoal para ajudar na análise e limpeza dos dados. 
import suportedata as sd
```

```python
# algumas configurações de visualização do DataFrame
pd.set_option('display.max_columns', None)
pd.set_option('display.max_colwidth', None)
```

```python
# importando a primeira visualização do DataFrame
clientes = pd.read_json('dados/Telco-Customer-Churn.json')
```

```python
clientes
```

Nesse primeiro momento é possível visualizar duas coisas, a primeira é que os dados se encontra agrupado em colunas, para isso é preciso extrair esses dados em um processo chamado de normalização. E a segunda é que o nosso DataFrame possui 7267 linhas e 6 colunas, essas 7267 linhas são exatamente as informações de todos os clientes que foram disponibilizados.

```python
# foram criados alguns DataFrames de apoio
customer = pd.json_normalize(clientes['customer']) 
phone = pd.json_normalize(clientes['phone']) 
internet = pd.json_normalize(clientes['internet']) 
account = pd.json_normalize(clientes['account'])

# esses DataFrames foram colocados em uma lista 
lista = [customer, phone, internet, account]

# dropando as colunas com informações já normalizadas. 
clientes_v01 = clientes.copy()
clientes_v01.drop(columns=['customer', 'phone','internet','account'], inplace=True)
```

Após o processo de normalização das colunas foram criados alguns DataFrames de apoio. Como vimos anteriormente, estes DataFrames carregam informações dos clientes e para melhor entender quais sãos essas informações, visualizaremos e identificaremos que  informações são essas que cada um desses novos DataFrames possuem.

# Juntando os DataFrames

### DataFrame Original

```python
clientes_v01.head()
```

Como é possível ver, cada um destes DataFrames secundários trazem informações diferentes sobre os mesmos usuário, sendo assim, juntá-los ao DataFrame original no qual somete ficaram as informações de **Churn** e os IDs dos clientes é fundamental para podermos fazer melhores analises de nossos dados.

### Novo DataFrame

```python
clientes_v02 = pd.concat([clientes_v01, customer, phone, internet, account], axis=1)
clientes_v02.head()
```

# Tradução das Variáveis

Antes de dar continuidade, farei uma tradução das variáveis(Colunas), de modo a facilitar nossa compreensão e também para evitar possíveis problemas na seleção das variáveis.

```python
variaveis = {
    'gender' : 'genero',
    'SeniorCitizen' : 'senior',
    'Partner' : 'conjuge',
    'Dependents' : 'dependentes',
    'tenure' : 'permanencia',
    'PhoneService' : 'ServicoTelefonico',
    'MultipleLines' : 'MultiplasLinhas',
    'InternetService': 'ServicoInternet',
    'OnlineSecurity' : 'SeguroOnline',
    'DeviceProtection' : 'ProtecaoDoDispositivo',
    'TechSupport' : 'SuporteTecnico',
    'StreamingMovies' : 'StreamingFilmes',
    'Contract' : 'contrato',
    'PaperlessBilling' : 'FaturaOnline',
    'PaymentMethod' : 'MetodoDePagamento',
    'Charges.Monthly' : 'GastoMensal',
    'Charges.Total' : 'GastoTotal'
}
valores = {
      'Female' : 'Mulher',
    'Male' : 'Homem',
    'Yes' : 'Sim',
    'No' : 'Não',
    'No phone service' : 'SemTelefone',
    'No internet service' : 'SemInternet',
    'Fiber optic' :'FibraOptica',
    'One year' : 'UmAno',
    'Two year' : 'DoisAnos',
    'Month-to-month' : 'Mensal',
    'Mailed check' : 'EnvioDeCheque',
    'Electronic check' : 'ChequeEletronico',
    'Credit card (automatic)' : 'CartaoCredito',
    'Bank transfer (automatic)' : 'TransferenciaBancaria'    
}


clientes_v03 = clientes_v02.rename(columns = variaveis)
clientes_v03.replace(to_replace = valores, inplace = True)
clientes_v03.head()
```

# Verificando as Variáveis

Agora que já temos o nosso novo DataFrame para poder trabalhar, precisamos fazer a verificação dos valores e das inconsistências de cada variável, para que nossa análise seja a mais confiável possível.

```python
clientes_v02.info()
```

Em uma primeira análise podemos verificar que não existem elementos faltando em nenhuma das variáveis, porém falta olhar ainda as duas variáveis que ficaram no nosso DataFrame original, a variável `Churn` e a variável `customerID`. E existem outras inconsistências no nosso DataFrame como, por exemplo, a variável `Charges.Total` esta identificada como `object`, sendo que já havíamos visto anteriormente que ela se travada do valor acumulado gasto por um cliente, ou seja, ela deveria esta como `float` (número decimal).

## Verificando a variável `Churn`

```python
print('Variável Churn:')
print(clientes_v02['Churn'].unique())
```

```python
# query para verificar os elementos incorretos 
clientes_v03.query('Churn == ""')
```

Mesmo que em uma primeira análise não vimos elementos faltando em nenhuma das variáveis, na variável `Churn` existem 224 elementos “em branco”, são elementos que não foram preenchidos adequadamente. Neste ponto é importa analisar se esta falta de informação ira impactar em nossa análise. Tendo em vista que nosso objetivo é justamente criar um modelo para a prevenção dos cancelamentos, estes dados que estão preenchidos inadequadamente não nos beneficia em nada, podendo até mesmo nos prejudicar, sendo assim, optarei por removê-los do nosso DataFrame.

```python
# excluindo os elementos incorretos
clientes_v04 = clientes_v03[~(clientes_v02['Churn'] == '')]

# refazendo a query para verifica se os elementos errados ainda estão no DataFrame 
churn_nao_preenchidos = clientes_v04.query('Churn == ""').shape[0]

print(f'Total de não preenchidos na coluna Churn:{churn_nao_preenchidos}')
```

## Variável 'customerID'

```python
clientes_v04['customerID'].value_counts()
```

Outro ponto a se verificar do DataFrame original é a coluna `customerID`. Essa coluna é usada na identificação de cada cliente, sendo assim cada cliente possui um ID único, e em nosso estudo esta informação não ira nos acrescentar valor, sendo assim ela sera excluída de nossa DataFrame.

```python
clientes_v05 = clientes_v03.drop(columns=['customerID'], axis= 1)
clientes_v05.head()
```

## Variável `GastoTotal`

A variável `GastoTotal` deveria esta no formato `float`, mas como observamos anteriormente ela está sendo classificada como `object`. Sendo assim dando uma olhada um pouco melhor na variável 

```python
clientes_v05['GastoTotal'].value_counts()
```

Quando é feita uma contagem de valores que aparecem nesta variável notamos que exitem 11 valores que não foram preenchidos. Porem se olharmos melhor esses valores comparados a outras variáveis podemos ter um noção o motivo da falta destes valores.

```python
clientes_v05.query('GastoTotal == " "')[['Churn', 'genero','conjuge',
                                         'dependentes','permanencia',
                                         'GastoMensal', 'GastoTotal']]
```

Analisando simultaneamente as variáveis `GastoTotal` e `permanencia` é possível concluir que o  11 clientes não possuem valores de gastos, pois são clientes novos, ainda não completaram seu primeiro mês como cliente de nossa empresa. Sendo assim podemos prosseguir em trocar o tipo de valor dessa variável 

```python
# criando mais uma versão do DataFrame
clientes_v06 = clientes_v05.copy()

# Realizando a conversão.
clientes_v06['GastoTotal'] = pd.to_numeric(clientes_v06['GastoTotal'], errors='coerce')
```

```python
clientes_v06.info()
```

```python
clientes_v06.query('permanencia  == 0 ')
```

Agora que a variável possui uma nova tipagem, observamos que os valores que antes estavam vazios agora estão preenchidos como NaN (Not-A-Number), isso significa que não possui valores algum. Antes de solucionar esse problema é melhor verificamos se mais alguma das  variáveis do nosso DataFrame também possui valores como NaN.

```python
clientes_v06.isna().sum()
```

Como somente a variável `GastoTotal` possui elementos vazios e sabemos agora e esses elementos são justamente aqueles que ainda não fizeram nenhum pagamento não realizando gasto com empresa preencheremos a falta destes valores com 0 (zero).

```python
# Trocando os NaN por zeros.
clientes_v07 = clientes_v06.fillna(0)
```

```python
clientes_v07.query('GastoTotal == 0')[['Churn', 'genero','conjuge',
                                         'dependentes','permanencia',
                                         'GastoMensal', 'GastoTotal']]
```

# Criando a variável `GastoDiario`

Criarei uma variável novo somente para ver se ela poderá acrescentar algum valor em nossos estudos.
Esta variável trará o gasto de cada cliente diariamente, posteriormente poderemos ver se ela nos ajudar ou não. 

```python
clientes_v08 = clientes_v07.copy()
clientes_v08['GastoDiario'] = round(clientes_v08['GastoMensal'] /30,2) 
clientes_v08.head()
```
