```python
# Importando o pandas para fazer a análise 
import pandas as pd

# Importando o pacote pessoal para ajudar na análise e limpeza dos dados. 
import suportedata as sd
```

```python
# algumas configurações de visualização do DataFrame
pd.set_option('display.max_columns', None)
pd.set_option('display.max_colwidth', None)
```

```python
# importando a primeira visualização do DataFrame
clientes = pd.read_json('dados/Telco-Customer-Churn.json')
```

```python
clientes
```

Nesse primeiro momento é possível visualizar duas coisas, a primeira é que os dados se encontra agrupado em colunas, para isso é preciso extrair esses dados em um processo chamado de normalização. E a segunda é que o nosso DataFrame possui 7267 linhas e 6 colunas, essas 7267 linhas são exatamente as informações de todos os clientes que foram disponibilizados.

```python
# foram criados alguns DataFrames de apoio
customer = pd.json_normalize(clientes['customer']) 
phone = pd.json_normalize(clientes['phone']) 
internet = pd.json_normalize(clientes['internet']) 
account = pd.json_normalize(clientes['account'])

# esses DataFrames foram colocados em uma lista 
lista = [customer, phone, internet, account]

# dropando as colunas com informações já normalizadas. 
clientes_v01 = clientes.copy()
clientes_v01.drop(columns=['customer', 'phone','internet','account'], inplace=True)
```

Após o processo de normalização das colunas foram criados alguns DataFrames de apoio. Como vimos anteriormente, estes DataFrames carregam informações dos clientes e para melhor entender quais sãos essas informações, visualizaremos e identificaremos que  informações são essas que cada um desses novos DataFrames possuem.

# Verificando as Variáveis

Agora que já temos o nosso novo DataFrame para poder trabalhar, precisamos fazer a verificação dos valores e das inconsistências de cada variável, para que nossa análise seja a mais confiável possível.

```python
clientes_v02.info()
```

Em uma primeira análise podemos verificar que não existem elementos faltando em nenhuma das variáveis, porém falta olhar ainda as duas variáveis que ficaram no nosso DataFrame original, a variável `Churn` e a variável `customerID`. E existem outras inconsistências no nosso DataFrame como, por exemplo, a variável `Charges.Total` esta identificada como `object`, sendo que já havíamos visto anteriormente que ela se travada do valor acumulado gasto por um cliente, ou seja, ela deveria esta como `float` (número decimal).

## Verificando a variável `Churn`

```python
print('Variável Churn:')
print(clientes_v02['Churn'].unique())
```

```python
# query para verificar os elementos incorretos 
clientes_v02.query('Churn == ""')
```

Mesmo que em uma primeira análise não vimos elementos faltando em nenhuma das variáveis, na variável `Churn` existem 224 elementos “em branco”, são elementos que não foram preenchidos adequadamente. Neste ponto é importa analisar se esta falta de informação ira impactar em nossa análise. Tendo em vista que nosso objetivo é justamente criar um modelo para a prevenção dos cancelamentos, estes dados que estão preenchidos inadequadamente não nos beneficia em nada, podendo até mesmo nos prejudicar, sendo assim, optarei por removê-los do nosso DataFrame.

```python
# excluindo os elementos incorretos
clientes_v03 = clientes_v02[~(clientes_v02['Churn'] == '')]

# refazendo a query para verifica se os elementos errados ainda estão no DataFrame 
churn_nao_preenchidos = clientes_v03.query('Churn == ""').shape[0]

print(f'Total de não preenchidos na coluna Churn:{churn_nao_preenchidos}')
```
