# Normalizando as colunas Json

```python
# foram criados alguns DataFrames de apoio
customer = pd.json_normalize(clientes['customer']) 
phone = pd.json_normalize(clientes['phone']) 
internet = pd.json_normalize(clientes['internet']) 
account = pd.json_normalize(clientes['account'])

# esses DataFrames foram colocados em uma lista 
lista = [customer, phone, internet, account]

# dropando as colunas com informações já normalizadas. 
clientes_v01 = clientes.copy()
clientes_v01.drop(columns=['customer', 'phone','internet','account'], inplace=True)
```

Os dados estão em um formato JSON onde as informações estão em colunas, para isso é preciso normalizar, algo parecido com um processo de “extração”.

## Coluna`GastoTotal`

A coluna `Charges.Total`, agora chamada de `GastoTotal`, possui a tipagem de *_object_*, porém como visto inicialmente ela deveria ser uma coluna numérica.

```python
clientes_v05['GastoTotal'].value_counts()
```

Um dos motivos é que esta possui 11 elementos do qual seu valor está em branco.

```python
clientes_v05.query('GastoTotal == " "')[['Churn', 'genero','conjuge',
                                         'dependentes','permanencia',
                                         'GastoMensal', 'GastoTotal']]
```

Esses 11 valores são de clientes novos, que não completaram nem um mês de permanência na **Alura Voz**.

```python
# criando mais uma versão do DataFrame
clientes_v06 = clientes_v05.copy()

# Realizando a conversão.
clientes_v06['GastoTotal'] = pd.to_numeric(clientes_v06['GastoTotal'], errors='coerce')
```

```python
clientes_v06.info()
```

```python
clientes_v06.query('permanencia  == 0 ')[['Churn', 'genero','conjuge',
                                         'dependentes','permanencia',
                                         'GastoMensal', 'GastoTotal']]
```

Após troca da tipagem da coluna `GastoTotal` os 11 valores ficaram preenchidos como não numéricos.

```python
clientes_v06.isna().sum()
```

E como somente essa coluna possui valores não numéricos, é possível alterar seus valores, para zero, sem interferir em outras colunas.

```python
# trocando os NaN por zeros.
clientes_v07 = clientes_v06.fillna(0)
```

```python
clientes_v07.query('GastoTotal == 0')[['Churn', 'genero','conjuge',
                                         'dependentes','permanencia',
                                         'GastoMensal', 'GastoTotal']]
```

# Criando a variável `GastoDiario`

A coluna `GastoDiario` será criada de modo que se testa a possibilidade dela ajudar em futuras análises.

```python
# criando um novo DataFrame
clientes_v08 = clientes_v07.copy()

# criando a coluna GastoDiario
clientes_v08['GastoDiario'] = round(clientes_v08['GastoMensal'] /30,2) 
clientes_v08.head()
```
